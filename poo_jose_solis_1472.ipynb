#### conceptos de POO (programacion orientada a objetos) en python

el prompt en gemini AI

lenguaje Python, crear la clase "animal", con los atributos "nombre", "raza" crear el contructor animal, y las funciones ladrar(). correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
class Animal:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} (raza: {self.raza}) dice: ¡Guau, guau!😍")

    def correr(self):
        print(f"{self.nombre} está corriendo a toda velocidad.💕")

    def Morder(self):
      print(f"{self.nombre} está mordiendo a la vecina🌹")

# Creación de los objetos
pitbull = Animal("Rocky", "Pitbull")
chihuahua = Animal("Chispita", "Chihuahua")
perrocallejero = Animal("solovino", "chafa")

# Mostrar datos del primer perro
print(f"Datos del primer perro:\nNombre: {pitbull.nombre}\nRaza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()
pitbull.Morder()

print("-" * 20)

# Mostrar datos del segundo perro
print(f"Datos del segundo perro:\nNombre: {chihuahua.nombre}\nRaza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()

print("-" * 20)

# Mostrar datos del tercer perro
print(f"Datos del tercer perro:\nNombre: {perrocallejero.nombre}\nRaza: {perrocallejero.raza}")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()
print("programa realizado por jose solis NL 51")
```
